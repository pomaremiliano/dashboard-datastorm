## 1. ¿Cuál es el tipo de ruta ideal para operar con la mayor rentabilidad de combustible?

- Comparar CPK entre tipos de ruta (larga, media, corta) — ANOVA


```python
# Calcular el promedio de kilómetros por ruta
tabla_rutas["Km_Promedio_Ruta"] = (
    tabla_rutas["kmstotales"] / tabla_rutas["Cantidad de viajes"]
)


# Clasificar tipo de ruta según el promedio
def clasificar_ruta(km):
    if km < 150:
        return "Corta"
    elif 150 <= km <= 300:
        return "Media"
    else:
        return "Larga"


tabla_rutas["TipoRuta"] = tabla_rutas["Km_Promedio_Ruta"].apply(clasificar_ruta)


tabla_rutas[["Ruta", "Km_Promedio_Ruta", "TipoRuta"]].head()
```

ruta corta = 0 - 150km

ruta media = 150 - 300km

ruta larga = 300km en adelante


```python
print(tabla_rutas["TipoRuta"].unique())
```

```python
import scipy.stats as stats

cpk_larga = tabla_rutas[tabla_rutas["TipoRuta"] == "Larga"]["CPK"]
cpk_media = tabla_rutas[tabla_rutas["TipoRuta"] == "Media"]["CPK"]
cpk_corta = tabla_rutas[tabla_rutas["TipoRuta"] == "Corta"]["CPK"]
```

```python
# Prueba ANOVA
f_stat, p_value = stats.f_oneway(cpk_larga, cpk_media, cpk_corta)
print(f"Estadístico F: {f_stat}, Valor p: {p_value}")
if p_value < 0.05:
    print("Hay diferencias significativas entre al menos dos grupos de rutas.")
else:
    print("No hay diferencias significativas entre los grupos de rutas.")
```

```python
# Analizar el CPK promedio por tipo de ruta para determinar el tipo ideal
cpk_por_tipo_ruta = tabla_rutas.groupby("TipoRuta")["CPK"].mean().sort_values()
print("CPK promedio por tipo de ruta:")
print(cpk_por_tipo_ruta)

tipo_ruta_ideal = cpk_por_tipo_ruta.idxmin()
print(
    f"\nEl tipo de ruta ideal para operar con la mayor rentabilidad de combustible es: {tipo_ruta_ideal}"
)
```

- ANOVA entre tipos de unidad o comparación de medias. Como hay poca información entre unidades, si solo hay 2 diferentes entonces se aplicaría un t-test.


Clasificación de unidades:

- Tres quartiles de CPK por tipo de unidad
- Comparación de medias entre unidades


```python
df_gastos_unidad.info()
```

```python
# Calcular los cuantiles para los límites
q1 = df_gastos_unidad_peaje_sort["Kms Totales"].quantile(1 / 3)
q2 = df_gastos_unidad_peaje_sort["Kms Totales"].quantile(2 / 3)


def clasificar_unidad_quantiles(kms):
    if kms < q1:
        return "Poca actividad"
    elif kms < q2:
        return "Actividad media"
    else:
        return "Alta actividad"


df_gastos_unidad_peaje_sort["TipoUnidad"] = df_gastos_unidad_peaje_sort["Kms Totales"].apply(
    clasificar_unidad_quantiles
)
```

### Nomenclatura de los quantiles para clasificación de unidades

- **Poca actividad**: Unidades cuyo total de kilómetros recorridos está por debajo del primer tercil (q1).
- **Actividad media**: Unidades con kilómetros recorridos entre el primer tercil (q1) y el segundo tercil (q2).
- **Alta actividad**: Unidades cuyo total de kilómetros recorridos está por encima del segundo tercil (q2).

> **Nota:** Los terciles (quantiles 1/3 y 2/3) dividen la muestra en tres grupos de igual tamaño según la variable "Kms Totales".

## 3. ¿Qué se puede aprender de las 10 unidades más costosas por kilómetro?


```python
from scipy.stats import ttest_ind

# Seleccionar las 10 unidades con mayor CPK
top10 = df_gastos_unidad_peaje.nlargest(10, "CPK")

# Crear subconjunto con el resto de las unidades
resto = df_gastos_unidad_peaje[~df_gastos_unidad_peaje["Unidad"].isin(top10["Unidad"])]

# Extraer las columnas de CPK para cada grupo
cpk_top10 = top10["CPK"].dropna()
cpk_resto = resto["CPK"].dropna()

# Verificar que ambos grupos tengan suficientes datos
if len(cpk_top10) > 1 and len(cpk_resto) > 1:
    t_stat, p_value = ttest_ind(cpk_top10, cpk_resto, equal_var=False)
    print(f"Estadístico t: {t_stat:.4f}, Valor p: {p_value:.6f}")

    if p_value < 0.05:
        print(
            "Hay diferencias significativas en el CPK entre las 10 unidades más costosas y el resto."
        )
    else:
        print(
            "No hay diferencias significativas entre las 10 unidades más costosas y el resto."
        )
else:
    print("No hay suficientes datos para aplicar la prueba t.")
```

## 4. ¿Cómo influye la estacionalidad en la variabilidad del CPK en cada ruta?


```python
from scipy.stats import f_oneway

# Calcular CPK si no existe
if "CPK" not in df_merged_mantto.columns:
    df_merged_mantto["CPK"] = (
        df_merged_mantto["Costo por carga"] + df_merged_mantto["Gasto Mantenimiento"]
    ) / df_merged_mantto["kmstotales"].replace(0, pd.NA)
    df_merged_mantto["CPK"] = (
        df_merged_mantto["CPK"].replace([float("inf"), -float("inf")], pd.NA).fillna(0)
    )

# Asegurarse de que la columna de fecha es datetime
df_merged_mantto["Fecha de Transacción"] = pd.to_datetime(
    df_merged_mantto["Fecha de Transacción"], errors="coerce"
)

# Extraer el mes (1 a 12)
df_merged_mantto["Mes"] = df_merged_mantto["Fecha de Transacción"].dt.month

# Agrupar por mes y recolectar listas de CPK por mes
cpk_por_mes = [
    grupo["CPK"].dropna()
    for nombre, grupo in df_merged_mantto.groupby("Mes")
    if len(grupo["CPK"].dropna()) > 1  # asegurar que haya datos suficientes por mes
]

# Aplicar ANOVA si hay suficientes grupos
if len(cpk_por_mes) >= 2:
    f_stat, p_value = f_oneway(*cpk_por_mes)
    print(f"Estadístico F: {f_stat:.4f}, Valor p: {p_value:.6f}")

    if p_value < 0.05:
        print(
            "Hay diferencias significativas en el CPK entre meses (estacionalidad detectada)."
        )
    else:
        print("No hay diferencias significativas entre meses.")
else:
    print("No hay suficientes datos por mes para aplicar ANOVA.")
```

```python
# Obtener el promedio de CPK por mes y mostrar los meses con CPK más altos
cpk_mes = df_merged_mantto.groupby("Mes")["CPK"].mean().sort_values(ascending=False)
print("Meses con CPK promedio más alto:")
print(cpk_mes)
```

```python
import matplotlib.pyplot as plt

cpk_mes.plot(kind="bar", figsize=(10, 5), color="skyblue")
plt.title("Histograma de CPK promedio por mes")
plt.xlabel("Mes")
plt.ylabel("CPK promedio")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

El valor del estadístico F fue 41.35 y el valor p fue prácticamente 0.000000, por lo que se rechaza la hipótesis nula. Esto indica que existe una diferencia estadísticamente significativa en el CPK entre los distintos meses, lo que confirma la presencia de un comportamiento estacional.

## 5. ¿Qué características explican el bajo rendimiento en las rutas con mayor CPK?


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Seleccionar las variables relevantes
df_cluster = df_maestra_cluster[
    ["CPK", "Proyecto", "Unidad", "TipoRuta", "TipoUnidad", "Mes"]
].copy()

# Convertir variables categóricas a variables dummy
df_encoded = pd.get_dummies(df_cluster, drop_first=True)

# Normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

# Aplicar K-means con 3 clusters (puedes probar otros valores)
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded["Cluster"] = kmeans.fit_predict(X_scaled)

# Añadir etiquetas de cluster al original
df_cluster["Cluster"] = df_encoded["Cluster"]
```

```python
# Comparar promedio de CPK por cluster
cluster_summary = (
    df_cluster.groupby("Cluster")["CPK"].mean().sort_values(ascending=False)
)
print(cluster_summary)
```

```python
# Analizar qué características dominan en el cluster más caro
cluster_alto_cpk = df_cluster[df_cluster["Cluster"] == cluster_summary.index[0]]
resumen_caracteristicas = cluster_alto_cpk[
    ["Proyecto", "TipoRuta", "TipoUnidad"]
].value_counts(normalize=True)
print(resumen_caracteristicas.head(10))
```

```python
import matplotlib.pyplot as plt

# Tomar las 10 combinaciones más frecuentes
top_caracteristicas = resumen_caracteristicas.head(10)

# Crear etiquetas combinadas para el eje x
labels = [
    f"{proj}\n{ruta}\n{unidad}" for (proj, ruta, unidad) in top_caracteristicas.index
]

plt.figure(figsize=(10, 6))
plt.barh(labels, top_caracteristicas.values, color="tomato")
plt.xlabel("Proporción dentro del cluster más costoso")
plt.title("Características dominantes en el cluster más alto de CPK")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import itertools
import pandas as pd

# Modelo de escenarios para encontrar combinaciones que den el mínimo CPK posible

# 1. Seleccionar las combinaciones posibles de Proyecto, TipoRuta y TipoUnidad
proyectos = df_maestra_cluster["Proyecto"].unique()
tipos_ruta = df_maestra_cluster["TipoRuta"].unique()
tipos_unidad = df_maestra_cluster["TipoUnidad"].unique()


# 2. Crear todas las combinaciones posibles
combinaciones = list(itertools.product(proyectos, tipos_ruta, tipos_unidad))
df_escenarios = pd.DataFrame(combinaciones, columns=["Proyecto", "TipoRuta", "TipoUnidad"])

# 3. Calcular el CPK promedio observado para cada combinación real en los datos
cpk_promedio = (
    df_maestra_cluster.groupby(["Proyecto", "TipoRuta", "TipoUnidad"])["CPK"]
    .mean()
    .reset_index()
)

# 4. Unir los escenarios con los CPK promedio observados
df_escenarios = df_escenarios.merge(
    cpk_promedio, on=["Proyecto", "TipoRuta", "TipoUnidad"], how="left"
)

# 5. Filtrar solo las combinaciones que existen en los datos (CPK no nulo)
df_escenarios_validos = df_escenarios.dropna(subset=["CPK"])

# 6. Ordenar para encontrar las combinaciones con menor CPK
escenarios_min_cpk = df_escenarios_validos.sort_values(by="CPK").head(10)

print("Top 10 combinaciones de Proyecto, TipoRuta y TipoUnidad con menor CPK observado:")
display(escenarios_min_cpk)
```

```python
# Analizar las rutas más usadas por mes

# Asegurarse de que la columna de fecha es datetime y extraer el mes si no existe
if "Mes" not in df_merged.columns:
    df_merged["Fecha de Transacción"] = pd.to_datetime(df_merged["Fecha de Transacción"], errors="coerce")
    df_merged["Mes"] = df_merged["Fecha de Transacción"].dt.month

# Contar la cantidad de viajes por ruta y mes
rutas_mes = (
    df_merged.groupby(["Mes", "Ruta"])["ord_number"]
    .count()
    .reset_index()
    .rename(columns={"ord_number": "Cantidad de viajes"})
    .sort_values(["Mes", "Cantidad de viajes"], ascending=[True, False])
)

# Obtener el top 5 de rutas más usadas por cada mes
top_rutas_mes = rutas_mes.groupby("Mes").head(5)

print("Top 5 rutas más usadas por mes:")
display(top_rutas_mes)
```
