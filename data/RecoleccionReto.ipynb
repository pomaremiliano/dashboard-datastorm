```python
df_mantenimiento = pd.read_excel("./old/Gastos de mantenimientos por unidad Q1.xlsx")
df_mantenimiento.info()
```

```python
df_mantenimiento = df_mantenimiento.rename(columns={"UNIDAD": "Unidad"})
# Suma de gastos de mantenimiento por unidad
gastos_mantto = (
    df_mantenimiento.groupby("Unidad")["$ TOTAL"]
    .sum()
    .reset_index()
    .rename(columns={"$ TOTAL": "Gasto Mantenimiento"})
)
df_mantenimiento.head(10)
```

```python
df_merged["Costo por carga"] = pd.to_numeric(
    df_merged["Costo por carga"], errors="coerce"
)
df_merged["Costo por litro"] = pd.to_numeric(
    df_merged["Costo por litro"], errors="coerce"
)

gastos_combustible = (
    df_merged.groupby("Unidad")
    .agg(
        {
            "Litros": "sum",
            "Costo por carga": "sum",
            "Costo por litro": "mean",
            "kmstotales": "sum",
        }
    )
    .reset_index()
    .rename(
        columns={"Costo por carga": "Gasto Combustible", "kmstotales": "Kms Totales"}
    )
)
gastos_combustible
```

```python
df_gastos_unidad = gastos_combustible.merge(gastos_mantto, on="Unidad", how="left")

# Llenar NaN con 0 para unidades sin gastos de algún tipo
df_gastos_unidad.fillna(0, inplace=True)
```

```python
df_gastos_unidad
```

```python
df_gastos_unidad["CPK"] = (
    df_gastos_unidad["Gasto Combustible"] + df_gastos_unidad["Gasto Mantenimiento"]
) / df_gastos_unidad["Kms Totales"].replace(0, pd.NA)

df_gastos_unidad["CPK"] = (
    df_gastos_unidad["CPK"].replace([float("inf"), -float("inf")], pd.NA).fillna(0)
)
```

```python
df_gastos_unidad_sort = df_gastos_unidad.sort_values(by="CPK", ascending=False)
```

```python
# Crear una columna 'Ruta' combinando 'Ciudad Origen' y 'Ciudad Destino'
df_merged["Ruta"] = df_merged["Ciudad Origen"] + " → " + df_merged["Ciudad Destino"]

# Agrupar por 'Ruta' y sumar los kilómetros totales y litros consumidos
tabla_rutas = (
    df_merged.groupby("Ruta")
    .agg(
        {
            "kmstotales": "sum",
            "Litros": "sum",
            "ord_number": "count",  # cantidad de viajes/rutas
        }
    )
    .reset_index()
    .rename(columns={"ord_number": "Cantidad de viajes"})
    .sort_values(by="Cantidad de viajes", ascending=False)
)

tabla_rutas10 = tabla_rutas.head(10)
```

```python
num_rutas = df_merged["Ruta"].nunique()
print(f"Cantidad de rutas: {num_rutas}")
```

```python
# Crear una tabla de rutas por unidad, mostrando para cada unidad las rutas realizadas y el número de viajes
tabla_rutas_unidad = (
    df_merged.groupby(["Unidad", "Ruta"])
    .agg(
        {
            "kmstotales": "sum",
            "Litros": "sum",
            "ord_number": "count",  # cantidad de viajes/rutas por unidad
        }
    )
    .reset_index()
    .rename(columns={"ord_number": "Cantidad de viajes"})
    .sort_values(by=["Unidad", "Cantidad de viajes"], ascending=[True, False])
)


tabla_rutas_unidad10 = tabla_rutas_unidad.head(10)
```

```python
df_merged_mantto = df_merged.merge(
    gastos_mantto, left_on="Unidad", right_on="Unidad", how="left"
)


cpk_ruta = (
    df_merged_mantto.groupby("Ruta")
    .agg(
        {
            "Costo por carga": "sum",
            "Gasto Mantenimiento": "sum",
            "kmstotales": "sum",
            "ord_number": "count",
            "Litros": "sum",
            "Unidad": "nunique",  # número de unidades que realizaron la ruta
        }
    )
    .reset_index()
)

cpk_ruta["CPK"] = (
    (cpk_ruta["Costo por carga"] + cpk_ruta["Gasto Mantenimiento"])
    / cpk_ruta["kmstotales"].replace(0, pd.NA)
).fillna(0)

cpk_ruta = cpk_ruta.sort_values(by="CPK", ascending=False)
```

```python
# Calcular eficiencia de combustible (km por litro) por ruta
tabla_rutas["Eficiencia (km/l)"] = tabla_rutas["kmstotales"] / tabla_rutas["Litros"]

# Calcular CPK (Costo por kilómetro) por ruta
# Usamos cpk_ruta, que ya tiene el cálculo por ruta
cpk_dict = cpk_ruta.set_index("Ruta")["CPK"].to_dict()
tabla_rutas["CPK"] = tabla_rutas["Ruta"].map(cpk_dict)

tabla_rutas_eff = tabla_rutas.replace([float("inf"), -float("inf")], pd.NA).dropna(
    subset=["Eficiencia (km/l)"]
)

# Top 10 rutas más eficientes
top10_rutas_eficientes = tabla_rutas_eff.sort_values(
    by="Eficiencia (km/l)", ascending=False
).head(10)

# Top 10 rutas menos eficientes (descartando eficiencias negativas o cero)
top10_rutas_menos_eficientes = (
    tabla_rutas_eff[tabla_rutas_eff["Eficiencia (km/l)"] > 0]
    .sort_values(by="Eficiencia (km/l)")
    .head(10)
)

top10_rutas_eficientes.insert(0, "Posición", range(1, 11))
top10_rutas_menos_eficientes.insert(0, "Posición", range(1, 11))

top10_rutas_eficientes, top10_rutas_menos_eficientes
```

```python
# Calcular eficiencia de combustible (km por litro) por unidad y mantener CPK
df_gastos_unidad["Eficiencia (km/l)"] = (
    df_gastos_unidad["Kms Totales"] / df_gastos_unidad["Litros"]
)

# Seleccionar columnas relevantes para mostrar eficiencia y CPK
cols = ["Unidad", "Kms Totales", "Litros", "Eficiencia (km/l)", "CPK"]

# Top 10 unidades más eficientes (descartando eficiencias negativas o cero)
top10_unidades_eficientes = (
    df_gastos_unidad[df_gastos_unidad["Eficiencia (km/l)"] > 0]
    .sort_values(by="Eficiencia (km/l)", ascending=False)
    .head(10)
    .reset_index(drop=True)
)[cols]

# Top 10 unidades menos eficientes (descartando eficiencias negativas o cero)
top10_unidades_menos_eficientes = (
    df_gastos_unidad[df_gastos_unidad["Eficiencia (km/l)"] > 0]
    .sort_values(by="Eficiencia (km/l)")
    .head(10)
    .reset_index(drop=True)
)[cols]

top10_unidades_eficientes, top10_unidades_menos_eficientes
```

## 1. ¿Cuál es el tipo de ruta ideal para operar con la mayor rentabilidad de combustible?

- Comparar CPK entre tipos de ruta (larga, media, corta) — ANOVA


```python
# Calcular el promedio de kilómetros por ruta
tabla_rutas["Km_Promedio_Ruta"] = (
    tabla_rutas["kmstotales"] / tabla_rutas["Cantidad de viajes"]
)


# Clasificar tipo de ruta según el promedio
def clasificar_ruta(km):
    if km < 150:
        return "Corta"
    elif 150 <= km <= 300:
        return "Media"
    else:
        return "Larga"


tabla_rutas["TipoRuta"] = tabla_rutas["Km_Promedio_Ruta"].apply(clasificar_ruta)


tabla_rutas[["Ruta", "Km_Promedio_Ruta", "TipoRuta"]].head()
```

ruta corta = 0 - 150km 

ruta media = 150 - 300km

ruta larga = 300km en adelante

```python
print(tabla_rutas["TipoRuta"].unique())
```

```python
import scipy.stats as stats

cpk_larga = tabla_rutas[tabla_rutas["TipoRuta"] == "Larga"]["CPK"]
cpk_media = tabla_rutas[tabla_rutas["TipoRuta"] == "Media"]["CPK"]
cpk_corta = tabla_rutas[tabla_rutas["TipoRuta"] == "Corta"]["CPK"]
```

```python
# Prueba ANOVA
f_stat, p_value = stats.f_oneway(cpk_larga, cpk_media, cpk_corta)
print(f"Estadístico F: {f_stat}, Valor p: {p_value}")
if p_value < 0.05:
    print("Hay diferencias significativas entre al menos dos grupos de rutas.")
else:
    print("No hay diferencias significativas entre los grupos de rutas.")
```

```python
# Analizar el CPK promedio por tipo de ruta para determinar el tipo ideal
cpk_por_tipo_ruta = tabla_rutas.groupby("TipoRuta")["CPK"].mean().sort_values()
print("CPK promedio por tipo de ruta:")
print(cpk_por_tipo_ruta)

tipo_ruta_ideal = cpk_por_tipo_ruta.idxmin()
print(f"\nEl tipo de ruta ideal para operar con la mayor rentabilidad de combustible es: {tipo_ruta_ideal}")
```

```python
# Mostrar las 10 rutas con menor CPK (más rentables por kilómetro)
top10_rutas_menor_cpk = tabla_rutas.sort_values(by="CPK").head(10)
top10_rutas_menor_cpk[["Ruta", "CPK", "Eficiencia (km/l)", "kmstotales", "Cantidad de viajes"]]
```

## 2. ¿Cómo se puede utilizar la distribución del CPK por tipo de unidad para identificar unidades más eficientes?

- ANOVA entre tipos de unidad o comparación de medias. Como hay poca información entre unidades, si solo hay 2 diferentes entonces se aplicaría un t-test.


Clasificación de unidades:

- Tres quartiles de CPK por tipo de unidad
- Comparación de medias entre unidades


```python
df_gastos_unidad.info()
```

```python
# Calcular los cuantiles para los límites
q1 = df_gastos_unidad["Kms Totales"].quantile(1 / 3)
q2 = df_gastos_unidad["Kms Totales"].quantile(2 / 3)


def clasificar_unidad_quantiles(kms):
    if kms < q1:
        return "Poca actividad"
    elif kms < q2:
        return "Actividad media"
    else:
        return "Alta actividad"


df_gastos_unidad["TipoUnidad"] = df_gastos_unidad["Kms Totales"].apply(
    clasificar_unidad_quantiles
)
```

### Nomenclatura de los quantiles para clasificación de unidades

- **Poca actividad**: Unidades cuyo total de kilómetros recorridos está por debajo del primer tercil (q1).
- **Actividad media**: Unidades con kilómetros recorridos entre el primer tercil (q1) y el segundo tercil (q2).
- **Alta actividad**: Unidades cuyo total de kilómetros recorridos está por encima del segundo tercil (q2).

> **Nota:** Los terciles (quantiles 1/3 y 2/3) dividen la muestra en tres grupos de igual tamaño según la variable "Kms Totales".

```python
df_gastos_unidad.head(50)
```

```python
print(df_gastos_unidad["TipoUnidad"].value_counts())
```

```python
# Prueba de ANOVA
# Agrupar el CPK por tipo de unidad usando la columna 'TipoUnidad' de df_gastos_unidad
cpk_baja = df_gastos_unidad[df_gastos_unidad["TipoUnidad"] == "Poca actividad"]["CPK"]
cpk_media = df_gastos_unidad[df_gastos_unidad["TipoUnidad"] == "Actividad media"]["CPK"]
cpk_alta = df_gastos_unidad[df_gastos_unidad["TipoUnidad"] == "Alta actividad"]["CPK"]

from scipy import stats

# Verificar que cada grupo tenga al menos 2 valores
if (len(cpk_baja) > 1) and (len(cpk_media) > 1) and (len(cpk_alta) > 1):
    f_stat, p_value = stats.f_oneway(cpk_baja, cpk_media, cpk_alta)
    print(f"Estadístico F: {f_stat}, Valor p: {p_value}")

    if p_value < 0.05:
        print("Hay diferencias significativas entre al menos dos tipos de unidades.")
    else:
        print("No hay diferencias significativas entre los grupos.")
else:
    print(
        "No hay suficientes datos en cada grupo para realizar ANOVA. Cada grupo necesita al menos 2 valores."
    )
```

```python
# Mostrar las 10 unidades con menor CPK (más rentables por kilómetro)
top10_unidades_menor_cpk = df_gastos_unidad[df_gastos_unidad["CPK"] > 0].sort_values(by="CPK").head(10)
top10_unidades_menor_cpk[["Unidad", "CPK", "Kms Totales", "Litros", "Eficiencia (km/l)"]]
```

```python
# Analizar eficiencia y CPK promedio por tipo de unidad
resumen_tipo_unidad = df_gastos_unidad.groupby("TipoUnidad").agg(
    eficiencia_promedio=("Eficiencia (km/l)", "mean"),
    cpk_promedio=("CPK", "mean"),
    cantidad_unidades=("Unidad", "count")
).sort_values(by="eficiencia_promedio", ascending=False)

print("Resumen por tipo de unidad:")
print(resumen_tipo_unidad)

mejor_tipo_eficiencia = resumen_tipo_unidad["eficiencia_promedio"].idxmax()
mejor_tipo_cpk = resumen_tipo_unidad["cpk_promedio"].idxmin()

print(f"\nEl mejor tipo de unidad en eficiencia (km/l) es: {mejor_tipo_eficiencia}")
print(f"El mejor tipo de unidad en CPK (más bajo) es: {mejor_tipo_cpk}")
```

## 3. ¿Qué se puede aprender de las 10 unidades más costosas por kilómetro?

```python
from scipy.stats import ttest_ind

# Seleccionar las 10 unidades con mayor CPK
top10 = df_gastos_unidad.nlargest(10, 'CPK')

# Crear subconjunto con el resto de las unidades
resto = df_gastos_unidad[~df_gastos_unidad['Unidad'].isin(top10['Unidad'])]

# Extraer las columnas de CPK para cada grupo
cpk_top10 = top10['CPK'].dropna()
cpk_resto = resto['CPK'].dropna()

# Verificar que ambos grupos tengan suficientes datos
if len(cpk_top10) > 1 and len(cpk_resto) > 1:
    t_stat, p_value = ttest_ind(cpk_top10, cpk_resto, equal_var=False)
    print(f"Estadístico t: {t_stat:.4f}, Valor p: {p_value:.6f}")

    if p_value < 0.05:
        print("Hay diferencias significativas en el CPK entre las 10 unidades más costosas y el resto.")
    else:
        print("No hay diferencias significativas entre las 10 unidades más costosas y el resto.")
else:
    print("No hay suficientes datos para aplicar la prueba t.")
```

```python
# Mostrar las 10 unidades con mayor CPK (más costosas por kilómetro)
top10_unidades_mas_costosas = df_gastos_unidad[df_gastos_unidad["CPK"] > 0].sort_values(by="CPK", ascending=False).head(10)
top10_unidades_mas_costosas[["Unidad", "CPK", "Kms Totales", "Litros", "Eficiencia (km/l)"]]
```

El análisis de comparación de medias arrojó un estadístico t = 3.6331 y un valor p = 0.0054. Como p < 0.05, se rechaza la hipótesis nula, concluyendo que las 10 unidades con mayor CPK presentan una eficiencia significativamente distinta al resto de la flota.

## 4. ¿Cómo influye la estacionalidad en la variabilidad del CPK en cada ruta?

```python
from scipy.stats import f_oneway

# Calcular CPK si no existe
if "CPK" not in df_merged_mantto.columns:
    df_merged_mantto["CPK"] = (
        (df_merged_mantto["Costo por carga"] + df_merged_mantto["Gasto Mantenimiento"])
        / df_merged_mantto["kmstotales"].replace(0, pd.NA)
    )
    df_merged_mantto["CPK"] = (
        df_merged_mantto["CPK"].replace([float("inf"), -float("inf")], pd.NA).fillna(0)
    )

# Asegurarse de que la columna de fecha es datetime
df_merged_mantto['Fecha de Transacción'] = pd.to_datetime(df_merged_mantto['Fecha de Transacción'], errors='coerce')

# Extraer el mes (1 a 12)
df_merged_mantto['Mes'] = df_merged_mantto['Fecha de Transacción'].dt.month

# Agrupar por mes y recolectar listas de CPK por mes
cpk_por_mes = [
    grupo['CPK'].dropna()
    for nombre, grupo in df_merged_mantto.groupby('Mes')
    if len(grupo['CPK'].dropna()) > 1  # asegurar que haya datos suficientes por mes
]

# Aplicar ANOVA si hay suficientes grupos
if len(cpk_por_mes) >= 2:
    f_stat, p_value = f_oneway(*cpk_por_mes)
    print(f"Estadístico F: {f_stat:.4f}, Valor p: {p_value:.6f}")
    
    if p_value < 0.05:
        print("Hay diferencias significativas en el CPK entre meses (estacionalidad detectada).")
    else:
        print("No hay diferencias significativas entre meses.")
else:
    print("No hay suficientes datos por mes para aplicar ANOVA.")
```

```python
# Obtener el promedio de CPK por mes y mostrar los meses con CPK más altos
cpk_mes = df_merged_mantto.groupby('Mes')['CPK'].mean().sort_values(ascending=False)
print("Meses con CPK promedio más alto:")
print(cpk_mes)
```

```python
import matplotlib.pyplot as plt

cpk_mes.plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Histograma de CPK promedio por mes')
plt.xlabel('Mes')
plt.ylabel('CPK promedio')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

El valor del estadístico F fue 41.35 y el valor p fue prácticamente 0.000000, por lo que se rechaza la hipótesis nula. Esto indica que existe una diferencia estadísticamente significativa en el CPK entre los distintos meses, lo que confirma la presencia de un comportamiento estacional.

```python
# Crear tabla maestra para análisis de cluster
# Merge entre gastos_combustible, gastos_mantto, tabla_rutas_unidad, y unidad_info

# Primero, unimos gastos_combustible con gastos_mantto para tener CPK y mantenimiento
df_maestra = gastos_combustible.merge(gastos_mantto, on="Unidad", how="left")

# Agregamos TipoUnidad desde df_gastos_unidad (ya tiene CPK y TipoUnidad)
df_maestra = df_maestra.merge(
    df_gastos_unidad[["Unidad", "CPK", "TipoUnidad"]], on="Unidad", how="left"
)

# Crear unidad_info desde df_merged (o df_merged_mantto) con Proyecto y Mes
unidad_info = df_merged[["Unidad", "Proyecto", "Mes"]].drop_duplicates()

# Agregamos Proyecto y Mes desde unidad_info
df_maestra = df_maestra.merge(unidad_info, on="Unidad", how="left")

# Para TipoRuta, tomamos la ruta más frecuente por unidad desde tabla_rutas_unidad
tipo_ruta_por_unidad = (
    tabla_rutas_unidad
    .merge(tabla_rutas[["Ruta", "TipoRuta"]], on="Ruta", how="left")
    .groupby("Unidad")["TipoRuta"]
    .agg(lambda x: x.value_counts().idxmax())
    .reset_index()
)
df_maestra = df_maestra.merge(tipo_ruta_por_unidad, on="Unidad", how="left")

# Seleccionamos solo las columnas solicitadas y eliminamos duplicados
df_maestra_cluster = df_maestra[["Unidad", "CPK", "Proyecto", "TipoRuta", "TipoUnidad", "Mes"]].drop_duplicates()

df_maestra_cluster.head()
```

## 5. ¿Qué características explican el bajo rendimiento en las rutas con mayor CPK?


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Seleccionar las variables relevantes
df_cluster = df_maestra_cluster[['CPK', 'Proyecto', 'Unidad', 'TipoRuta', 'TipoUnidad', 'Mes']].copy()

# Convertir variables categóricas a variables dummy
df_encoded = pd.get_dummies(df_cluster, drop_first=True)

# Normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

# Aplicar K-means con 3 clusters (puedes probar otros valores)
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded['Cluster'] = kmeans.fit_predict(X_scaled)

# Añadir etiquetas de cluster al original
df_cluster['Cluster'] = df_encoded['Cluster']
```

```python
# Comparar promedio de CPK por cluster
cluster_summary = df_cluster.groupby('Cluster')['CPK'].mean().sort_values(ascending=False)
print(cluster_summary)
```

```python
# Analizar qué características dominan en el cluster más caro
cluster_alto_cpk = df_cluster[df_cluster['Cluster'] == cluster_summary.index[0]]
resumen_caracteristicas = cluster_alto_cpk[['Proyecto', 'TipoRuta', 'TipoUnidad']].value_counts(normalize=True)
print(resumen_caracteristicas.head(10))
```

### Interpretación de los clusters y sus resultados

En el análisis de clusters realizado, cada cluster representa un grupo de unidades/rutas con características y patrones de operación similares, identificados a partir de variables como CPK, tipo de ruta, tipo de unidad, proyecto y mes.

- **Cluster 0:** Generalmente agrupa unidades/rutas con CPK bajo y alta eficiencia. Suele estar compuesto por unidades de "Alta actividad" y rutas largas, lo que indica operaciones más rentables y eficientes.

- **Cluster 2:** Es el grupo con el **mayor CPK promedio** (más costoso por kilómetro). Este cluster destaca por:
    - Predominio de proyectos como LIVERPOOL, SAYER, WALMAR VH y CIRCUITOS.
    - Presencia de rutas largas y cortas, pero con unidades de "Actividad media" o "Poca actividad".
    - Combinaciones frecuentes como "LIVERPOOL – Ruta corta – Unidad de actividad media".

**Conclusión:**  
El cluster 2 identifica las combinaciones menos rentables, permitiendo enfocar acciones de mejora en esos proyectos, rutas y tipos de unidades. Los clusters ayudan a segmentar la flota y las operaciones para tomar decisiones específicas según el perfil de cada grupo.
