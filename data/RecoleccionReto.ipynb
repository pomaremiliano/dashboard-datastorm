```python
df_3 = pd.read_excel("./old/Ordenes Oct-Marzo 2025 Tec 2da version.xlsx")
df_3.info()
```

```python
df_3['Tracto'] = df_3['Tracto'].astype(str).str.strip()
df_C['Unidad'] = df_C['Unidad'].astype(str).str.strip()

# Merge: Combina usando "Tracto" y "Unidad"
df_merged = pd.merge(df_3, df_C, left_on='Tracto', right_on='Unidad', how='inner')

# Ahora puedes analizar el consumo por tracto
df_consumo_por_tracto = df_merged.groupby('Tracto')['Litros'].sum().reset_index()
df_consumo_por_tracto.columns = ['Tracto', 'Total Litros Consumidos']
df_consumo_por_tracto
```

```python
df_merged.info()
```

```python
df_mantenimiento = pd.read_excel("./old/Gastos de mantenimientos por unidad Q1.xlsx")
df_mantenimiento.info()
```

```python
df_mantenimiento = df_mantenimiento.rename(columns={'UNIDAD': 'Unidad'})
# Suma de gastos de mantenimiento por unidad
gastos_mantto = (
    df_mantenimiento
    .groupby('Unidad')['$ TOTAL'].sum()
    .reset_index()
    .rename(columns={'$ TOTAL': 'Gasto Mantenimiento'})
)
df_mantenimiento.head(10)
```

```python
df_merged['Costo por carga'] = pd.to_numeric(df_merged['Costo por carga'], errors='coerce')
df_merged['Costo por litro'] = pd.to_numeric(df_merged['Costo por litro'], errors='coerce')

gastos_combustible = (
    df_merged
    .groupby('Unidad')
    .agg({
        'Litros': 'sum',
        'Costo por carga': 'sum',
        'Costo por litro': 'mean',
        'kmstotales': 'sum'
    })
    .reset_index()
    .rename(columns={
        'Costo por carga': 'Gasto Combustible',
        'kmstotales': 'Kms Totales'
    })
)
gastos_combustible
```

```python
df_gastos_unidad = gastos_combustible.merge(
    gastos_mantto, on='Unidad', how='left'
)

# Llenar NaN con 0 para unidades sin gastos de algún tipo
df_gastos_unidad.fillna(0, inplace=True)
```

```python
df_gastos_unidad
```

```python
df_gastos_unidad['CPK'] = (
    (df_gastos_unidad['Gasto Combustible'] + df_gastos_unidad['Gasto Mantenimiento']) /
    df_gastos_unidad['Kms Totales'].replace(0, pd.NA)
)

df_gastos_unidad['CPK'] = df_gastos_unidad['CPK'].replace([float('inf'), -float('inf')], pd.NA).fillna(0)
```

```python
df_gastos_unidad = df_gastos_unidad.head(10)
```

```python
df_gastos_unidad_sort = df_gastos_unidad.sort_values(by='CPK', ascending=False)
```

```python
# Crear una columna 'Ruta' combinando 'Ciudad Origen' y 'Ciudad Destino'
df_merged['Ruta'] = df_merged['Ciudad Origen'] + " → " + df_merged['Ciudad Destino']

# Agrupar por 'Ruta' y sumar los kilómetros totales y litros consumidos
tabla_rutas = (
    df_merged
    .groupby('Ruta')
    .agg({
        'kmstotales': 'sum',
        'Litros': 'sum',
        'ord_number': 'count',# cantidad de viajes/rutas
    })
    .reset_index()
    .rename(columns={'ord_number': 'Cantidad de viajes'})
    .sort_values(by='Cantidad de viajes', ascending=False)
)

tabla_rutas10 = tabla_rutas.head(10)
```

```python
num_rutas = df_merged['Ruta'].nunique()
print(f"Cantidad de rutas: {num_rutas}")
```

```python
# Crear una tabla de rutas por unidad, mostrando para cada unidad las rutas realizadas y el número de viajes
tabla_rutas_unidad = (
    df_merged
    .groupby(['Unidad', 'Ruta'])
    .agg({
        'kmstotales': 'sum',
        'Litros': 'sum',
        'ord_number': 'count'  # cantidad de viajes/rutas por unidad
    })
    .reset_index()
    .rename(columns={'ord_number': 'Cantidad de viajes'})
    .sort_values(by=['Unidad', 'Cantidad de viajes'], ascending=[True, False])
)



tabla_rutas_unidad10 = tabla_rutas_unidad.head(10)
```

```python
df_merged_mantto = df_merged.merge(
    gastos_mantto, left_on='Unidad', right_on='Unidad', how='left'
)


cpk_ruta = (
    df_merged_mantto
    .groupby('Ruta')
    .agg({
        'Costo por carga': 'sum',
        'Gasto Mantenimiento': 'sum',
        'kmstotales': 'sum',
        'ord_number': 'count',
        'Litros': 'sum',
        'Unidad': 'nunique'  # número de unidades que realizaron la ruta
        
    })
    .reset_index()
)

cpk_ruta['CPK'] = (
    (cpk_ruta['Costo por carga'] + cpk_ruta['Gasto Mantenimiento']) /
    cpk_ruta['kmstotales'].replace(0, pd.NA)
).fillna(0)

cpk_ruta = cpk_ruta.sort_values(by='CPK', ascending=False)
```

```python
# Calcular eficiencia de combustible (km por litro) por ruta
tabla_rutas['Eficiencia (km/l)'] = tabla_rutas['kmstotales'] / tabla_rutas['Litros']

# Calcular CPK (Costo por kilómetro) por ruta
# Usamos cpk_ruta, que ya tiene el cálculo por ruta
cpk_dict = cpk_ruta.set_index('Ruta')['CPK'].to_dict()
tabla_rutas['CPK'] = tabla_rutas['Ruta'].map(cpk_dict)

tabla_rutas_eff = tabla_rutas.replace([float('inf'), -float('inf')], pd.NA).dropna(subset=['Eficiencia (km/l)'])

# Top 10 rutas más eficientes
top10_rutas_eficientes = tabla_rutas_eff.sort_values(by='Eficiencia (km/l)', ascending=False).head(10)

# Top 10 rutas menos eficientes (descartando eficiencias negativas o cero)
top10_rutas_menos_eficientes = tabla_rutas_eff[tabla_rutas_eff['Eficiencia (km/l)'] > 0].sort_values(by='Eficiencia (km/l)').head(10)

top10_rutas_eficientes.insert(0, 'Posición', range(1, 11))
top10_rutas_menos_eficientes.insert(0, 'Posición', range(1, 11))

top10_rutas_eficientes, top10_rutas_menos_eficientes
```

```python
# Calcular eficiencia de combustible (km por litro) por unidad y mantener CPK
df_gastos_unidad['Eficiencia (km/l)'] = df_gastos_unidad['Kms Totales'] / df_gastos_unidad['Litros']

# Seleccionar columnas relevantes para mostrar eficiencia y CPK
cols = ['Unidad', 'Kms Totales', 'Litros', 'Eficiencia (km/l)', 'CPK']

# Top 10 unidades más eficientes (descartando eficiencias negativas o cero)
top10_unidades_eficientes = (
    df_gastos_unidad[df_gastos_unidad['Eficiencia (km/l)'] > 0]
    .sort_values(by='Eficiencia (km/l)', ascending=False)
    .head(10)
    .reset_index(drop=True)
)[cols]

# Top 10 unidades menos eficientes (descartando eficiencias negativas o cero)
top10_unidades_menos_eficientes = (
    df_gastos_unidad[df_gastos_unidad['Eficiencia (km/l)'] > 0]
    .sort_values(by='Eficiencia (km/l)')
    .head(10)
    .reset_index(drop=True)
)[cols]

top10_unidades_eficientes, top10_unidades_menos_eficientes
```

```python
with pd.ExcelWriter("Rutas_Resumen.xlsx") as writer:
    df_gastos_unidad_sort.to_excel(writer, sheet_name="Gastos por Unidad", index=False),
    tabla_rutas10.to_excel(writer, sheet_name="Rutas", index=False)
    tabla_rutas_unidad10.to_excel(writer, sheet_name="Rutas_Unidad", index=False)
    cpk_ruta.to_excel(writer, sheet_name="CPK_Ruta", index=False)
    top10_rutas_eficientes.to_excel(writer, sheet_name="Top 10 Rutas Mas Eficientes", index=False)
    top10_rutas_menos_eficientes.to_excel(writer, sheet_name="Top 10 Rutas Menos Eficientes", index=False)
    top10_unidades_eficientes.to_excel(writer, sheet_name="Top 10 Unidades Más Eficientes", index=False)
    top10_unidades_menos_eficientes.to_excel(writer, sheet_name="Top 10 Unidades Menos Eficientes", index=False)

print("Exportación de rutas completada")
```
