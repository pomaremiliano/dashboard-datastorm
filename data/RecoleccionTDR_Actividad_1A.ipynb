```python
df_3 = pd.read_excel("./old/Ordenes Oct-Marzo 2025 Tec 2da version.xlsx")
df_3.info()
```

```python
df_3['Tracto'] = df_3['Tracto'].astype(str).str.strip()
df_C['Unidad'] = df_C['Unidad'].astype(str).str.strip()

# Merge: Combina usando "Tracto" y "Unidad"
df_merged = pd.merge(df_3, df_C, left_on='Tracto', right_on='Unidad', how='inner')

# Ahora puedes analizar el consumo por tracto
df_consumo_por_tracto = df_merged.groupby('Tracto')['Litros'].sum().reset_index()
df_consumo_por_tracto.columns = ['Tracto', 'Total Litros Consumidos']
df_consumo_por_tracto
```

```python
tabla_proyecto_flota = (
    df_merged
    .groupby(['ID Proyecto', 'Proyecto', 'Flota'])
    .agg({
        'Litros': 'sum',
        'Unidad': 'count',  # cantidad de registros de carga
        'kmstotales' : 'sum'
    })
    .reset_index())
```

```python
# Crear una columna 'Ruta' combinando 'Ciudad Origen' y 'Ciudad Destino'
df_merged['Ruta'] = df_merged['Ciudad Origen'] + " → " + df_merged['Ciudad Destino']

# Agrupar por 'Ruta' y sumar los kilómetros totales y litros consumidos
tabla_rutas = (
    df_merged
    .groupby('Ruta')
    .agg({
        'kmstotales': 'sum',
        'Litros': 'sum',
        'ord_number': 'count',# cantidad de viajes/rutas
    })
    .reset_index()
    .rename(columns={'ord_number': 'Cantidad de viajes'})
    .sort_values(by='Cantidad de viajes', ascending=False)
)

tabla_rutas10 = tabla_rutas.head(10)
```

```python
# Crear una tabla de rutas por unidad, mostrando para cada unidad las rutas realizadas y el número de viajes
tabla_rutas_unidad = (
    df_merged
    .groupby(['Unidad', 'Ruta'])
    .agg({
        'kmstotales': 'sum',
        'Litros': 'sum',
        'ord_number': 'count'  # cantidad de viajes/rutas por unidad
    })
    .reset_index()
    .rename(columns={'ord_number': 'Cantidad de viajes'})
    .sort_values(by=['Unidad', 'Cantidad de viajes'], ascending=[True, False])
)

tabla_rutas_unidad10 = tabla_rutas_unidad.head(10)
```

```python
with pd.ExcelWriter("Rutas_Resumen.xlsx") as writer:
    tabla_rutas10.to_excel(writer, sheet_name="Rutas", index=False)
    tabla_rutas_unidad10.to_excel(writer, sheet_name="Rutas_Unidad", index=False)

print("Exportación de rutas completada")
```

```python
consumo_por_proyecto = (
    df_merged
    .groupby('Proyecto')['Litros']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(consumo_por_proyecto['Proyecto'], consumo_por_proyecto['Litros'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Proyecto')
plt.ylabel('Litros Consumidos')
plt.title('Consumo de Combustible por Proyecto')
plt.tight_layout()
plt.show()
```

```python
consumo_por_tracto = (
    df_merged
    .groupby('Tracto')['Litros']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
top5_tractos = consumo_por_tracto.head(5)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top5_tractos['Tracto'], top5_tractos['Litros'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Tracto')
plt.ylabel('Litros Consumidos')
plt.title('Consumo de Combustible por Tracto')
plt.tight_layout()
plt.show()
```

```python
top5_tractos_lista = top5_tractos['Tracto'].tolist()

top5_tractos_proyecto = df_merged[df_merged['Tracto'].isin(top5_tractos_lista)]
```

```python
tracto_proyecto_litros = (
    top5_tractos_proyecto
    .groupby(['Tracto', 'Proyecto'])['Litros']
    .sum()
    .reset_index()
    .sort_values(by='Litros', ascending=False)
)
```

```python
tracto_proyecto_litros['Tracto_Proyecto'] = (
    tracto_proyecto_litros['Tracto'] + ' | ' + tracto_proyecto_litros['Proyecto']
)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(tracto_proyecto_litros['Tracto_Proyecto'], tracto_proyecto_litros['Litros'], color='teal')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Tracto | Proyecto')
plt.ylabel('Litros Consumidos')
plt.title('Consumo de Combustible por Proyecto (Top 5 Tractos)')
plt.tight_layout()
plt.show()
```

```python
kms_por_proyecto = (
    df_merged
    .groupby('Proyecto')['kmstotales']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(kms_por_proyecto['Proyecto'], kms_por_proyecto['kmstotales'], color='green')
plt.xticks(rotation=90)
plt.xlabel('Proyecto')
plt.ylabel('Kilómetros recorridos')
plt.title('Kms recorridos por Proyecto')
plt.tight_layout()
plt.show()
```

```python
df_merged['Mes'] = df_merged['Inicio de la Orden'].dt.to_period('M')
```

```python
kms_por_proyecto_mes = (
    df_merged
    .groupby(['Proyecto', 'Mes'])['kmstotales']
    .sum()
    .reset_index()
    .sort_values(['Mes', 'kmstotales'], ascending=[True, False])
)
```

```python
kms_por_proyecto_mes
```

```python
proyecto_top_por_mes = (
    kms_por_proyecto_mes
    .sort_values(['Mes', 'kmstotales'], ascending=[True, False])
    .groupby('Mes')
    .head(1)
    .reset_index(drop=True)
)
```

```python
proyecto_top_por_mes
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=proyecto_top_por_mes, x='Mes', y='kmstotales', hue='Proyecto', dodge=False)
plt.title('Proyecto con más km recorridos por mes')
plt.ylabel('Km Totales')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
consumo_proyecto_mes = (
    df_merged
    .groupby(['Proyecto', 'Mes'])['Litros']
    .sum()
    .reset_index()
    .sort_values(['Mes', 'Litros'], ascending=[True, False])
)
```

```python
consumo_proyecto_mes
```

```python
proyecto_top_consumo_por_mes = (
    consumo_proyecto_mes
    .sort_values(['Mes', 'Litros'], ascending=[True, False])
    .groupby('Mes')
    .head(2)
    .reset_index(drop=True)
)
```

```python
proyecto_top_consumo_por_mes
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=proyecto_top_consumo_por_mes, x='Mes', y='Litros', hue='Proyecto', dodge=False)
plt.title('Proyectos con mayor consumo de combustible al mes ')
plt.ylabel('Litros')
plt.xlabel('Mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
proyecto_mes_comb = pd.merge(
    kms_por_proyecto_mes,
    consumo_proyecto_mes,
    on=['Proyecto', 'Mes'],
    how='inner'  # solo proyectos que aparecen en ambos
)
```

```python
correlacion = proyecto_mes_comb[['kmstotales', 'Litros']].corr().iloc[0, 1]
print(f"Correlación entre km y litros: {correlacion:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=proyecto_mes_comb, x='kmstotales', y='Litros', hue='Mes', alpha=0.7)
plt.title(f'Dispersión de km vs litros por proyecto y mes (r = {correlacion:.2f})')
plt.xlabel('Kilómetros recorridos')
plt.ylabel('Litros consumidos')
plt.grid(True)
plt.tight_layout()
plt.show()
```
